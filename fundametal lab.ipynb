```python
import numpy as np

student_scores = np.loadtxt("/Students_data.csv",delimiter=",",skiprows=1,usecols=(1, 2, 3, 4))
subjects = ["Math", "Science", "English", "History"]

average_scores = np.mean(student_scores, axis=0)

highest_subject = subjects[np.argmax(average_scores)]

print("Average scores:", np.round(average_scores,2))
print("Subject with highest average score:", highest_subject)
```

```python
import numpy as np

# Load the dataset from CSV file
house_data = np.loadtxt(
    "/Home_Data.csv",
    delimiter=",",
    skiprows=1
)

# Create a boolean mask for houses with more than 4 bedrooms
mask = house_data[:, 0] > 4

# Extract sale prices of houses that match the condition
sale_prices = house_data[mask, 2]

# Calculate the average sale price
average_sale_price = np.mean(sale_prices)

print("Average sale price of houses with more than 4 bedrooms:", average_sale_price)
```

```python
import numpy as np

# Quarterly sales data (Q1, Q2, Q3, Q4)
sales_data = np.array([250000, 300000, 350000, 420000])

# Calculate total sales for the year
total_sales = np.sum(sales_data)

# Calculate percentage increase from Q1 to Q4
percentage_increase = ((sales_data[3] - sales_data[0]) / sales_data[0]) * 100

print("/total Sales_Data.csv", total_sales)
print("Percentage increase from Q1 to Q4:", percentage_increase, "%")
```

```python
import numpy as np

# Fuel efficiency data (in miles per gallon)
fuel_efficiency = np.array([18.5, 21.0, 24.5, 28.0, 32.0])

# Calculate average fuel efficiency
average_efficiency = np.mean(fuel_efficiency)

# Calculate percentage improvement from Model 1 to Model 5
percentage_improvement = (
    (fuel_efficiency[4] - fuel_efficiency[0]) / fuel_efficiency[0]
) * 100

print("/fuel_efficiency.csv", average_efficiency, "MPG")
print("Percentage improvement from Model 1 to Model 5:", percentage_improvement, "%")
```
